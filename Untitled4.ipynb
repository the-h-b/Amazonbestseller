# **Amazon.in web scraping demo**

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

 The headers parameter in the code provides a specific "User-Agent" value when making a request to a website. This value mimics a real web browser's user agent string, making the request appear as if it's coming from a regular browser rather than a script or bot. This can help avoid potential blocking or restrictions from websites that may treat automated requests differently.

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
webpage=requests.get('https://www.amazon.in/gp/bestsellers/apparel/ref=zg_bs_apparel_smy', headers =headers).text
```

soup is just a random variable

```python
soup = BeautifulSoup(webpage,'lxml')
```

in order to print out the well-formatted HTML structure of the web page content that you've scraped using Beautiful Soup. It's a useful way to see the HTML structure in an indented and human-readable format, making it easier to understand the hierarchy of elements and their relationships on the page.

```python
print(soup.prettify())
```

this code snippet helps to extract all the h1 tags from the page. in this case there's only one.

```python
h1_elements = soup.find_all('h1')
if h1_elements:
    h1_text = h1_elements[0].text
else:
    h1_text = "No <h1> element found"
print(h1_text)
```

This part extracts all the encessary informations and convert that information into a csv file

```python
final = pd.DataFrame()

for j in range(1, 10):
    webpage = requests.get('https://www.amazon.in/gp/bestsellers/apparel/ref=zg_bs_pg_2_apparel?ie=UTF8&pg={}'.format(j)).text
    soup = BeautifulSoup(webpage, 'lxml')
    #for link in soup.find_all('a'):
      #print(link.get('href'))


    companies = soup.find_all('div', class_='a-cardui _cDEzb_grid-cell_1uMOS expandableGrid p13n-grid-content')

    ranking = []
    title = []
    ratings = []
    reviews = []


    for company in companies:
        try:
            ranking.append(company.find('span', class_='zg-bdg-text').text.strip())
        except:
            ranking.append(np.nan)

        try:
            title.append(company.find('div', class_='_cDEzb_p13n-sc-css-line-clamp-3_g3dy1').text.strip())
        except:
            title.append(np.nan)

        try:
            reviews.append(company.find('span', class_='a-size-small').text.strip())
        except:
            reviews.append(np.nan)

        try:
            ratings.append(company.find('span', class_='a-icon-alt').text.strip())
        except:
            ratings.append(np.nan)


    df = pd.DataFrame({'ranking':ranking , 'title': title,  'reviews': reviews, 'ratings':ratings})
    final = pd.concat([final, df], ignore_index=True)

print(final)
final.to_csv('output.csv', index=False)
```
